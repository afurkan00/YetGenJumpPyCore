### Soru 4

```python
x = [1,2,3,4,5]
y = [5,4,3,2,1]
``` 

```python	
Output:
x = [5, 4, 3, 2, 1]
y = [1, 2, 3, 4, 5]
```

````python
x = [1,2,3,4,5]
y = [5,4,3,2,1]
print("x =", y)
print("y =", x)
````

````python
x = [1,2,3,4,5]
y = [5,4,3,2,1]

x , y = y, x #x i y'ye ata y kısmını da x'e ata demek bu kısım 
print("x", x)
print("y",y)
````

````python
x = [1,2,3,4,5]
y = [5,4,3,2,1]
temp = x
x = y
y = temp  
#ilk olarak x'i bir şeye atıyoruz. Sonrasında y'yi x'e atadık ve y kısmınıda temp'e atadık.
#temp kısmını kullanırken sağ ya da solda olup olmaması önemli.
````

````python
totalCandies = int(input("Toplam şeker sayısını giriniz: "))
personNum = int(input("Kişi sayısını giriniz: "))
leftCandies = totalCandies % personNum
print(f"Parçalanacak şeker sayısı: {leftCandies}")
````
