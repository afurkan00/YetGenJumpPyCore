Lütfen numarasını öğrenmek istediğiniz kişinin adını girin: Ahmet

````python
telefon_rehberi = {
    "Ahmet": "0532 678 13 19",
    "Mehmet": "0543 123 45 67",
    "Ayşe": "0555 987 65 43",
    "Fatma": "0566 234 56 78"
}

kisi = input('Telefonuna erişmek istediğiniz kişinin ismini giriniz:')
if kisi in telefon_rehberi:
    print(kisi + " isimli kişinin numarası şu şekildedir: " + telefon_rehberi[kisi])
else:
    print(kisi + " isimli kişi rehberde bulunamadı.")
````

# Problem 3

3 öğrenciden oluşan bir öğrenci not sözlüğü oluşturun. Bu sözlükte öğrencilerin notları value olarak bir listede toplansın.

Kullanıcıya hangi öğrencinin notlarını görmek istediğini sorun. Öğrencinin notu görüntülendiğinde program sonunda şöyle bir çıktı elde etmelisiniz:

```
Lütfen notlarını görmek istediğiniz öğrencinin adını girin: Mehmet

Mehmet isimli öğrencinin      1.Sınav Notu:72
                              2.Sınav Notu:66
                              3.Sınav Notu:48
Not Ortalaması: 62.0


````python
ogrenci_notlari = {
    "Mehmet": [72, 66, 48],
    "Ayşe": [85, 90, 78],
    "Ahmet": [60, 75, 70]
}

isim = input("Lütfen notlarını görmek istediğiniz öğrencinin adını girin: ")

if isim in ogrenci_notlari:
    notlar = ogrenci_notlari[isim]
    ortalama = sum(notlar) / len(notlar) #len(notlar) listenin içinde kaç eleman olduğunu sayar.Yani ortalamayı kaça böleceğimizi rahat şekilde ayarlamamızı sağlar.
    print(f"{isim} isimli öğrencinin not ortalaması {ortalama}")
else:
    print(f"{isim} isimli öğrenci bulunamadı.")
````
